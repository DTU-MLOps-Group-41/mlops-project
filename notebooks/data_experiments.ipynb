```python
import kagglehub
import os
import pandas as pd

path = kagglehub.dataset_download("tobiasbueck/multilingual-customer-support-tickets", force_download=True)
print(f"Dataset downloaded to: {path}")
files = os.listdir(path)
print(f"Files found: {files}")
```

```python
full_data_file = "aa_dataset-tickets-multi-lang-5-2-50-version.csv"
medium_data_file = "dataset-tickets-multi-lang-4-20k.csv"
small_data_file = "dataset-tickets-multi-lang3-4k.csv"

df_full = pd.read_csv(os.path.join(path, full_data_file))
df_medium = pd.read_csv(os.path.join(path, medium_data_file))
df_small = pd.read_csv(os.path.join(path, small_data_file))

print(f"Full data shape: {df_full.shape}")
print(f"Medium data shape: {df_medium.shape}")
print(f"Small data shape: {df_small.shape}")
```

```python
df_full["body"] = df_full["body"].astype(pd.StringDtype())
df_full["priority"] = df_full["priority"].astype("category")
df_full.dtypes
```
